```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix
import seaborn as sns
```

```python
df = pd.read_csv('whitewines.csv')
```

```python
df.columns
```

```python
df['quality'].unique()
```

```python
df['quality'].describe()
```

```python
%matplotlib inline
#df_attr=df.iloc[:,0:12]
sns.pairplot(df, diag_kind='kde', hue = 'quality')
```

```python
df['quality'].value_counts()
```

```python
def isTasty(quality):
    if quality >= 7:
        return 1
    else:
        return 0
```

```python
df['tasty'] = df['quality'].apply(isTasty)
```

```python
df.columns
```

```python
df['tasty'].value_counts()
```

```python
data = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
#target = df['tasty']
target = df['quality']
```

```python
data_train, data_test, target_train, target_test = train_test_split(data,target,test_size = 0.33,random_state=123)
```

```python
[subset.shape for subset in [data_train,data_test,target_train,target_test]]
```

## Training Classifiers

```python
simpleTree = DecisionTreeClassifier(max_depth=5)
```

```python
simpleTree.fit(data_train,target_train)
```

```python
gbmTree = GradientBoostingClassifier()
```

```python
gbmTree.fit(data_train,target_train)
```

```python
bgcl = BaggingClassifier(n_estimators=50, max_samples=.7 , oob_score=False)
bgcl = bgcl.fit(data_train,target_train)
#print(bgcl.oob_score_)
```

```python
rfTree = RandomForestClassifier()
```

```python
rfTree.fit(data_train,target_train)
```

```python
print("simpleTree ", simpleTree.score(data_test,target_test))
print("gbmTree ",gbmTree.score(data_test,target_test))
print("bgcl ", bgcl.score(data_test,target_test))
print("rfTree", rfTree.score(data_test,target_test))
```

```python
print('Confusion Matrix for simple, gradient boosted, and random forest tree classifiers:')
print('Simple Tree:\n',confusion_matrix(target_test,simpleTree.predict(data_test)),'\n')
print('Gradient Boosted:\n',confusion_matrix(target_test,gbmTree.predict(data_test)),'\n')
print('Random Forest:\n',confusion_matrix(target_test,rfTree.predict(data_test)))
```

```python
print('Feature Importances for GBM tree\n')
for importance,feature in zip(gbmTree.feature_importances_,['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']):
    print('{}: {}'.format(feature,importance))
```

```python
print('Feature Importances for rf tree\n')
for importance,feature in zip(rfTree.feature_importances_,['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']):
    print('{}: {}'.format(feature,importance))
```

## GridSearch  For RandomForest

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_estimators':[25,50,75,100], 'max_features':['auto', 'sqrt', 'log2']}
```

```python
wqp_clf = GridSearchCV(rfTree, param_grid, cv=5, scoring = 'accuracy')
```

```python
wqp_clf.fit(data_train,target_train)
```

```python
print(wqp_clf.best_params_)
```
